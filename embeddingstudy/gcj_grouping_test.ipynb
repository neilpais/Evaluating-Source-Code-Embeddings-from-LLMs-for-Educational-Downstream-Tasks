```python
#take 10 rows of each problem_id
def sample_per_problem(df, n=10):
    return df.groupby('problem_id').apply(lambda x: x.sample(n=min(len(x), n))).reset_index(drop=True)
df_sampled = sample_per_problem(df, n=10)
df_sampled =df_sampled.head(70)
df_sampled.to_csv("sampled_per_problem.csv", index=False)
```
