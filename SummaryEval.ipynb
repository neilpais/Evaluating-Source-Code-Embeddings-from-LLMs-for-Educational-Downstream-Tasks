```python
import pandas as pd
import numpy as np
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from bert_score import score
```

```python
df = pd.read_csv('/home/profniggastein/PycharmProjects/ReducedDataset/remotescripts/results/Reduced_localsample_df.csv')
```

```python


def calculate_bleu(reference, candidate):
    reference = reference.split()
    candidate = candidate.split()
    return sentence_bleu([reference], candidate, smoothing_function=SmoothingFunction().method1)


df['bleu_score'] = df.apply(
    lambda row: calculate_bleu(row['GenSummary'], row['summary']) if pd.notnull(row['summary']) and pd.notnull(
        row['GenSummary']) else None,
    axis=1
)
df['BERT_score'] = df.apply(
    lambda row: score([row["GenSummary"]], [row["summary"]], model_type="bert-large-uncased")[2][0].item(),
    axis=1
)
df
```

```python
df.to_csv("/home/profniggastein/PycharmProjects/ReducedDataset/remotescripts/results/Reduced_localsample_df_Evaluated.csv", index=False)
```
