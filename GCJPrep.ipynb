```python
import pandas as pd
```

```python
gcj_df = pd.read_csv("/home/profniggastein/PycharmProjects/ReducedDataset/gcj2009.csv")
gcj_df = gcj_df[gcj_df['file'].str.endswith('.cpp', na=False)]
selected_columns = ['username', 'task', 'file', 'flines', 'solution']
gcj_df = gcj_df[selected_columns].reset_index(drop=True)
gcj_df
```

```python
gcj_correct= gcj_df[gcj_df['solution'] == 1].reset_index(drop=True)
gcj_correctexp = gcj_correct.head(1500)
gcj_correct
```

```python
gcj_incorrect = gcj_df[gcj_df['solution'] == 0].reset_index(drop=True)
gcj_incorrect = gcj_incorrect.head(13768)
gcj_incorrectexp = gcj_incorrect.head(1500)
gcj_incorrect
```

```python
gcj_balanced = pd.concat([gcj_correct, gcj_incorrect], ignore_index=True)
gcj_balancedexp = pd.concat([gcj_correctexp, gcj_incorrectexp], ignore_index=True)
gcj_balanced = gcj_balanced.sample(frac=1).reset_index(drop=True)  # Shuffle the DataFrame
gcj_balancedexp = gcj_balancedexp.sample(frac=1).reset_index(drop=True)  # Shuffle the DataFrame
#gcj_balanced.to_csv('/home/profniggastein/PycharmProjects/ReducedDataset/gcj2009_balanced.csv', index=False)
gcj_balancedexp.to_csv('/home/profniggastein/PycharmProjects/ReducedDataset/gcj2009_balanced_exp.csv', index=False)
gcj_balanced
```
