```python
import os

pnum = 0
pnumstr = "{:05}".format(pnum)
totalp = 4052
extracted_df = pd.DataFrame()
while (pnum < totalp):
    curr_problem_path = '/home/profniggastein/PycharmProjects/ReducedDataset/Project_CodeNet/metadata/p' + pnumstr + '.csv'
    problem_folder_path = '/home/profniggastein/PycharmProjects/ReducedDataset/Project_CodeNet_C++1000/p' + pnumstr
    df = pd.read_csv(curr_problem_path)
    #if problem folder does not exist, skip
    if not os.path.exists(problem_folder_path):
        print(f"Problem folder does not exist: {problem_folder_path}")
        pnum += 1
        pnumstr = "{:05}".format(pnum)
        continue

    Cpp_chunk_df = df[df['filename_ext'] == 'cpp']
    C_chunk_df = df[df['filename_ext'] == 'c']

    frames = [extracted_df, C_chunk_df, Cpp_chunk_df]
    print(pnumstr + 'added')
    extracted_df = pd.concat(frames)
    pnum += 1
    pnumstr = "{:05}".format(pnum)

extracted_df.reset_index(inplace=True)
#extracted_df.to_csv('/home/profniggastein/PycharmProjects/ReducedDataset/mischeck_df.csv')
extracted_df
```
